```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import pandas as pd
```

```python
import os
print(os.listdir("./images"))
```

```python
def plot_3(data):
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3,2, figsize = (13,15))
    ax1.imshow(data[0][...,::-1])
    ax2.imshow(data[1][...,::-1])
    ax3.imshow(data[2][...,::-1])
    ax4.imshow(data[3][...,::-1])
    ax5.imshow(data[4][...,::-1])
    ax6.imshow(data[5][...,::-1])
    ax1.axis('off')
    ax2.axis('off')
    ax3.axis('off')
    ax4.axis('off')
    ax5.axis('off')
    ax6.axis('off')
```

```python
img_name = ['test_lstm_1.png', 'test_lstm_2.png', 
            'test_lr_1.png', 'test_lr_2.png', 'test_rf_1.png', 'test_rf_2.png']
```

```python
images = []
for name in img_name:
    path = './images/' + name
    img = cv2.imread(path)  
    images.append(img)
```

```python
plot_3(images)
```

```python
res_name = ['residual_1_1.png', 'residual_2_1.png']
```

```python
res_im = []
for name in res_name:
    path = './images/' + name
    img = cv2.imread(path)  
    res_im.append(img)
```

```python
def plot_2(data):
    fig, ((ax1, ax2)) = plt.subplots(1,2, figsize = (20,20))
    ax1.imshow(data[0][...,::-1])
    ax2.imshow(data[1][...,::-1])
    ax1.axis('off')
    ax2.axis('off')
```

```python
plot_2(res_im)
```
